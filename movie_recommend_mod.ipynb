```python
# All Important libraries to import
import numpy as np
import pandas as pd
import ast
import nltk
from nltk.stem import WordNetLemmatizer,PorterStemmer
import pickle
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
display(movies.head(2))
display(credits.head(2))
```

```python
print('Shape of movies df',movies.shape)
print('Shape of credits df',credits.shape)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(2)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(2)
```

```python
# Droping Na rows
movies.dropna(inplace=True)
```

```python
movies.isna().sum()
```

```python
#Checking duplicate records
movies.duplicated().sum()
```

```python
def fetch_name(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
# Extracting name of genres
movies['genres'] = movies['genres'].apply(fetch_name)
movies.head()
```

```python
# Extracting name of keywords
movies['keywords'] = movies['keywords'].apply(fetch_name)
movies.head()
```

```python
def fetch_actor(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(fetch_actor)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head(5)
```

```python
def join(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(join)
movies['crew'] = movies['crew'].apply(join)
movies['genres'] = movies['genres'].apply(join)
movies['keywords'] = movies['keywords'].apply(join)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies['tags'].head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: x.lower())
new.head()
```

```python
#Steming of words
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]   
    for i in text.split():
        y.append(ps.stem(i))
    return (' '.join(y))    
```

```python
new['tags']=new['tags'].apply(stem)
new.head()
```

# Model creation using  bag of world

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
vector
```

```python
# Most frequent words
cv.get_feature_names()
```

```python
vector.shape
```

```python
# For calculating distance b/w 2 vectors we are not using euclidian dist instead v are using cosine(angle b/w 2 vectors) dist
# and if value of cosine is less that means the vectors are quite similar
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    # here we are extracting top 5 similar movies
    for i in distances[1:5]: 
        print(new.iloc[i[0]].title)
```

### With the help of recommend function we will get recommended movies 

```python
recommend('Avatar')
```

```python
# Dumping model
pickle.dump(new,open('movie_list.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```
